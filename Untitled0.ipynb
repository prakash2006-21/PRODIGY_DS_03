```python
#import neccesary libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

#read csv datafiles
d=pd.read_csv('/content/drive/MyDrive/bank_data.csv')

# ... (Your existing code to preprocess 'd')

# Create a copy of the preprocessed DataFrame 'd' and name it 'd_new'
d_new = d.copy()

# Continue with the rest of your code...
data_new = d_new.copy()

# Encode categorical variables in the new dataset
label_encoders_new = {}
for column in data_new.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    data_new[column] = le.fit_transform(data_new[column])
    label_encoders_new[column] = le

# Define features and target in the new dataset
X_new = data_new.drop('deposit', axis=1)
y_new = data_new['deposit']

# Split the new dataset into training and testing sets
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(X_new, y_new, test_size=0.2, random_state=42)

# Train a decision tree classifier on the new dataset
clf_new = DecisionTreeClassifier(criterion="entropy",max_depth=3)
clf_new.fit(X_train_new, y_train_new)

# Predict on the test set of the new dataset
y_pred_new = clf_new.predict(X_test_new)

# Evaluate the model on the new dataset
accuracy_new = accuracy_score(y_test_new, y_pred_new)
report_new = classification_report(y_test_new, y_pred_new)

accuracy_new, report_new
```
